```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def cleanData(data):
    columns_to_drop = ['PassengerId','Name','Ticket','Cabin','Embarked']
    cleaned_data = data.drop(columns=columns_to_drop)
    # Imputation --> Replacing the NaN values with mean or median of that column.
    cleaned_data = cleaned_data.fillna(cleaned_data['Age'].mean())
    cleaned_data['Sex'] = le.fit_transform(cleaned_data['Sex'])
    return cleaned_data
```

```python
# Entropy Function 
def Entropy(column):
    counts = np.unique(column,return_counts=True)
    N = column.shape[0]
    entropy = 0.0
    for ix in counts[1]:
        p = ix/N
        entropy += (-1.0*p*np.log2(p))
    return entropy
```

```python
def divideData(xdata,col_name,threshold_value):
    right_data = pd.DataFrame([],columns=xdata.columns)
    left_data = pd.DataFrame([],columns = xdata.columns)
    
    for ix in range(xdata.shape[0]):
        value = xdata[col_name].loc[ix]
        
        if value > threshold_value:
            right_data = right_data.append(xdata.loc[ix])
        else:
            left_data = left_data.append(xdata.loc[ix])
        
    return right_data,left_data        
```

```python
def informationGain(xdata,col_name,threshold_value):
    right,left = divideData(xdata,col_name=col_name,threshold_value=threshold_value)
    r = float(right.shape[0])/xdata.shape[0]
    l = float(left.shape[0])/xdata.shape[0]
    
    if right.shape[0] == 0 or left.shape[0] == 0:
        return -10000000
    info_gain = Entropy(xdata.Survived) - (l*Entropy(left.Survived)+r*Entropy(right.Survived))
    return info_gain
```

```python
input_cols = ['Pclass',"Sex","Age","SibSp","Parch","Fare"]
output_cols = ["Survived"]

X = cleaned_data[input_cols]
Y = cleaned_data[output_cols]

print(X.shape,Y.shape)
```

```python
for fx in X.columns:
    print(fx)
    print(informationGain(cleaned_data,fx,cleaned_data[fx].mean()))
```

```python
class DecisionTree:
    def __init__(self,current_depth = 0,max_depth = 5):
        self.left = None
        self.right = None
        self.col_name = None
        self.threshold_value = None
        self.current_depth = current_depth
        self.max_depth = max_depth
        self.target = None
    
    def train(self,xdata):
        features = ['Pclass','Sex','Age','SibSp', 'Parch', 'Fare']
        info_gains = []
        
        for ix in features:
            i_gain = informationGain(xdata,col_name=ix,threshold_value=xdata[ix].mean())
            info_gains.append(i_gain)
        
        self.col_name = features[np.argmax(info_gains)]
        self.threshold_value = xdata[self.col_name].mean()
        print('Making Tree Feature is :',self.col_name)
        
        # split the data
        right_data,left_data = divideData(xdata,col_name=self.col_name,threshold_value=self.threshold_value)
        right_data = right_data.reset_index(drop = True)
        left_data = left_data.reset_index(drop=True)
        
        # leaf Node
        if right_data.shape[0] == 0 or left_data.shape[0] == 0:
            if xdata.Survived.mean()>=0.5:
                self.target = 'Survive'
            else:
                self.target = 'Dead'
            return
        # stop early if current_depth>=max_depth
        if self.current_depth>=self.max_depth:
            if xdata.Survived.mean()>=0.5:
                self.target = 'Survive'
            else:
                self.target = 'Dead'
            return
        # recursive case
        self.right = DecisionTree(current_depth = self.current_depth+1,max_depth = self.max_depth)
        self.right.train(right_data)
        self.left = DecisionTree(current_depth = self.current_depth+1,max_depth = self.max_depth)
        self.left.train(left_data)
        if xdata.Survived.mean()>=0.5:
            self.target = 'Survive'
        else:
            self.target = 'Dead'
        return
        
    def predict(self,test_data):
        if test_data[self.col_name]>self.threshold_value:
            # go to right
            #check for leaf node
            if self.right is None:
                return self.target
            return self.right.predict(test_data)
        else:
            if self.left is None:
                return self.target
            return self.left.predict(test_data)
```

```python
d = DecisionTree()
d.train(cleaned_data)
```
