```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_samples=800,n_features=2,centers=2,random_state=None)
print(X.shape,Y.shape)
```

## z = W.T*X

## Y_pred = sigmoid(z)

## Loss = -(1.0/m)*summission over m(Y(i)*log(Y_pred(i)) + (1-Y(i))*log(1-Y_pred(i)))

## W(j) = W - learning_rate*summision over m(Y_pred(i) - Y(i))*X(i)(j)

```python
import matplotlib.pyplot as plt
plt.figure(0)
plt.style.use('seaborn')
plt.xlabel('X0')
plt.ylabel('X1')
plt.scatter(X[:,0],X[:,1],c = Y,cmap = plt.cm.Accent)
plt.show()
```

```python
# Function that calculates sigmoid on Z
def sigmoid(z):
    return 1.0/(1+np.exp((-1.0)*z))

# Function that returns Y_pred
def predict(X,weights):
    """X -> mx(n+1) matrix, w -->(nX1,) vector"""
    z = np.dot(X,weights)
    predictions = sigmoid(z)
    return predictions

# Function to calculate loss
def loss(X,Y,weights):
    """Binary Cross Entropy"""
    Y_ = predict(X,weights)
    cost = np.mean(-Y*np.log(Y_) - (1-Y)*np.log(1-Y_))
    return cost

# Function to update weights.
def update(X,Y,weights,learning_rate):
    """Perform weight updates for 1 epoch"""
    Y_ = predict(X,weights)
    dw = np.dot(X.T,Y_ - Y)
    
    m = X.shape[0]
    weights = weights - learning_rate*dw/(float(m))
    return weights

# Function to train the dataset.
# MaxEpochs means maxIterations.
def train(X,Y,learning_rate=0.5,maxEpochs=1000):
    
    #Modify the input to handle the bias term
    ones = np.ones((X.shape[0],1))
    X = np.hstack((ones,X))
    
    weights = np.zeros(X.shape[1]) #n+1 entires
    
    
    #Iterate over all epochs and make updates
    
    for epoch in range(maxEpochs):
        
        weights = update(X,Y,weights,learning_rate)
        
        if epoch%10==0:
            l = loss(X,Y,weights)
            print("Epoch %d Loss %.4f"%(epoch,l))
            print(weights)
    return weights
```

```python
weights= train(X,Y)
print(weights)
```

### My line equation should be x0*w0 + x1*w1 +x2*w2 = 0 where x0 is always 1.
### Therefore, i initialised x1
### x2 = -(x0*w0 + x1*w1)/w2

```python
x1 = np.linspace(-6,8)
print(x1)
x2 = -(weights[0] + weights[1]*x1)/weights[2]
print(x2)
```

```python
plt.figure(0)
plt.style.use('seaborn')
plt.xlabel('X0')
plt.ylabel('X1')
plt.scatter(X[:,0],X[:,1],c = Y,cmap = plt.cm.Accent)
plt.plot(x1,x2,c = 'red')
plt.show()
```

```python
def getPredictions(X_Test,weights,labels=True):
    if X_Test.shape[1] != weights.shape[0]:
        ones = np.ones((X_Test.shape[0],1))
        X_Test = np.hstack((ones,X_Test))

    probs = predict(X_Test,weights)
    
    if not labels:
        return probs
    else:
        labels = np.zeros(probs.shape)
        labels[probs>=0.5] = 1
        return labels
```

```python
# Find the accuracy!
accuracy = []
Y_ = getPredictions(X,weights,labels=True)
training_acc = np.sum(Y_==Y)/Y.shape[0]
print(training_acc)
accuracy.append(training_acc)
```

```python
from sklearn.datasets import make_circles
from sklearn.datasets import make_moons

x1,y1 = make_circles(n_samples=800,random_state=None)
x2,y2 = make_moons(n_samples=800,random_state=None)
print(x1.shape,y1.shape)
print(x2.shape,y2.shape)
```

```python
circle_wts = train(x1,y1)
```

```python
Y_ = getPredictions(x1,circle_wts,labels=True)
training_acc = np.sum(Y_==y1)/y1.shape[0]
print(training_acc)
accuracy.append(training_acc)
```

```python
x1_ = np.linspace(-2,2,10)
x2_ = -(circle_wts[0] + circle_wts[1]*x1_)/circle_wts[2]
plt.figure(0)
plt.style.use('seaborn')
plt.xlabel('X0')
plt.ylabel('X1')
plt.scatter(x1[:,0],x1[:,1],c = y1,cmap = plt.cm.Accent)
plt.plot(x1_,x2_,c = 'red')
plt.show()
```

```python
moons_wts = train(x2,y2)
```

```python
Y_ = getPredictions(x2,moons_wts,labels=True)
training_acc = np.sum(Y_==y2)/y2.shape[0]
print(training_acc)
accuracy.append(training_acc)
```

```python
x1_ = np.linspace(-2,2,10)
x2_ = -(moons_wts[0] + moons_wts[1]*x1_)/moons_wts[2]
plt.figure(0)
plt.style.use('seaborn')
plt.xlabel('X0')
plt.ylabel('X1')
plt.scatter(x2[:,0],x2[:,1],c = y2,cmap = plt.cm.Accent)
plt.plot(x1_,x2_,c = 'red')
plt.show()
```
